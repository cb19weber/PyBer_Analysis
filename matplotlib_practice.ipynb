```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Module 5.1.4 Skill Drill - use Object-Oriented Interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker="D", linewidth=2)
ax.legend('Boston')
ax.set_xlabel('Date')
ax.set_ylabel('Fares ($)')
ax.set_ylim(0,50)
ax.grid()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="Purple", label="Boston")
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fares ($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
# Module 5.1.6 Skill Drill!
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
plt.scatter(x, y, color='Green', marker='x', s=x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare vs Total Rides')
```

```python
plt.scatter(y_axis, x_axis, color='Red', label="Chicago", s=[data*3 for data in y_axis])
plt.xlabel('Fare ($')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[data*5 for data in y_axis], color="skyblue", alpha=.8, edgecolors='black', linewidths=2, label="Boston")
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.invert_yaxis()
ax.set_xlim(0,50)
```

```python
categories = ['pants','shirts','outerwear','pajamas','underwear','shoes']
revenue = [24,15,5,20,10,25]
explode = [0,.1,0,0,0,0]
plt.pie(revenue, labels=categories, autopct="%1.1f%%", explode=explode)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold",
          "indianred", "tomato", "mistyrose"]
explode_values = [0,0,.2,0,0,0,.2,0,0,0,0,0]
ax.pie(y_axis, labels=x_axis, autopct="%.1f%%", shadow=True, startangle=-90, colors=colors, explode=explode_values)
ax.invert_yaxis()
plt.show()
```
